### Import the libraries

```python
# Import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Check if basic functionality works

```python
def check_setup():
    # Create a simple DataFrame
    df = pd.DataFrame({
        'Numbers': [1, 2, 3, 4],
        'Squares': [1, 4, 9, 16]
    })

    print("Pandas DataFrame:\n", df)

    # Create a simple numpy array and calculate its mean
    arr = np.array([1, 2, 3, 4, 5])
    print("Numpy Array:\n", arr)
    print("Mean of Numpy Array:", np.mean(arr))

    # Plot a basic graph using matplotlib
    plt.plot(df['Numbers'], df['Squares'], marker='o')
    plt.title("Plot of Numbers and their Squares")
    plt.xlabel("Numbers")
    plt.ylabel("Squares")
    plt.grid(True)
    plt.show()
```

### Run the function to check functionality

```python
check_setup()
```
