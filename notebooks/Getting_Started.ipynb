### **Example 2**

```python
p = 100.0
i = 0.08
n = 4
m = 12

def f(p, i, n, m):
  return p * (1 + i/m)**(n*m)

print(jax.grad(f, argnums=(0))(p, i, n, m))
```

### **Example 3**

```python
i_s = jnp.linspace(0.01, 0.15, 100)
ys = jax.vmap(jax.grad(f, argnums=(0)), in_axes=(None, 0, None, None))(p, i_s, n, m)

plt.plot(i_s, ys)
plt.title('Marginal Return', loc='left', size=14)
plt.xlabel('Interest Rate')
plt.show()
```
