### **Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### **Read In the Dataset**

```python
df = pd.read_excel(io='/content/business-analytics/datasets/Household_Debt_and_Credit.xlsx',
                   sheet_name='Page 3 Data',
                   skiprows=3)
```

### **Visualize the Dataset**

```python
df.head()
```

### **Rename a Column**

```python
df = df.rename(columns={'Unnamed: 0' : 'Quarter'})
df.head()
```

### **Plotting**

```python
mortage_timeline = df['Mortgage']
#mortage_timeline = df.Mortgage
plt.plot(mortage_timeline)
plt.show()
```

### **Convert to Time Data Type**

```python
df['Quarter'] = df['Quarter'].apply(lambda x: '20'+x[:2]+x[3:])
df['Quarter'] = pd.PeriodIndex(df['Quarter'], freq='Q').astype('datetime64[ns]')
```

### **Plotting**

```python
mortage_timeline = df.Mortgage
timeline = df.Quarter
plt.plot(timeline, mortage_timeline)

# Don't worry about this
from matplotlib.ticker import MaxNLocator
plt.gca().xaxis.set_major_locator(MaxNLocator(nbins=6))  # Set maximum of 6 ticks on the x-axis
plt.show()
```
