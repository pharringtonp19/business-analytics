#### **Function**

```python
def game(home_score, away_score):

  if home_score > away_score:
    print("The Home team won!")

  elif home_score < away_score:
    print("The Away team won!")

  else:
    print("Tie")
```

```python
h_score = 4
a_score = 3

game(h_score, a_score)
```

```python
h_score = 2
a_score = 2
game(h_score, a_score)
```

```python
h_score = 1
a_score = 2
game(h_score, a_score)
```
