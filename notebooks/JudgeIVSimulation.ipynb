```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

```python
key = jax.random.PRNGKey(0)
n = 5000
```

```python
def foreclosed(case_type, judge_leniency):
  return jnp.where(case_type > judge_leniency, 1.0, 0.0)


judge_leniency = jnp.linspace(0, 1, 100)
k1, _ = jax.random.split(key)
case_types = jax.random.uniform(k1, shape=(n,))
```

```python
for i in case_types[:10]:
  ys = jax.vmap(foreclosed, in_axes=(None, 0))(i, judge_leniency)
  plt.plot(judge_leniency, ys)
plt.show()
```
