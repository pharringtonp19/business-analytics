```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

```python
# Generate random uniform data
data = jax.random.uniform(jax.random.PRNGKey(1), shape=(5, 10))

# Normalize by column
data = data / jnp.sum(data, axis=0)

# Visualize the normalized data
plt.imshow(data)

# Add colorbar
plt.colorbar()

# Set custom y-axis ticks
plt.yticks(ticks=range(5), labels=['Loan Modification', 'Short Sale', 'Arms Length Sale', 'Foreclosure', 'Other'])

plt.xticks(ticks=range(10))
plt.xlabel('Judge', size=14)
plt.title('Court Outcome', loc='left')

# Show the plot
plt.show()
```
