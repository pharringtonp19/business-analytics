### **Import Libraries**

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

### **Get Company Data**

```python
# List of companies
list_of_companies = ['AAPL', 'TSLA']

# Start Date
start_date = '2020-03-01'

# End Data
end_date = '2024-10-01'

# Initialize an empty dictionary to collect data
data = {}

# Loop over each company in the list
for company in list_of_companies:

    # Download the stock data
    stock_data = yf.download(company, start=start_date, end=end_date)


    data[company] = stock_data['Close'].values.reshape(-1,) # Collecting 'High' data


df = pd.DataFrame(data, index=stock_data.index)
```

### **Plot**

```python
for company in list_of_companies:
  plt.plot(df[company], label=company)


### ADDITIONAL PLOTTING CODE TO IMPROVE THE FIGURE
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.title('Stock Price', loc='left', size=14)
plt.legend(frameon=False)
plt.show()
```
