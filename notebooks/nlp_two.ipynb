### **Import Libraries**

```python
from transformers import pipeline
import matplotlib.pyplot as plt
```

### **Configure Model**

```python
fill_mask = pipeline("fill-mask", model='distilbert-base-uncased')
```

### **Sentence**

```python
sentence = "The two fundamental aspects of programming are data and [MASK]."
```

### **Predictions**

```python
predictions = fill_mask(sentence)
```

### **Plot**

```python
tokens = [prediction['token_str'] for prediction in predictions]
scores = [prediction['score'] for prediction in predictions]
plt.figure(figsize=(8, 6))
plt.barh(tokens, scores, color="#36454F")
plt.xlabel("Score", size=14)
plt.title("Top Predictions for the Masked Word", loc='left')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest score on top
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```
