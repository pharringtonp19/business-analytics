### **Import Libraries**

```python
import numpy as np
from functools import partial
import matplotlib.pyplot as plt
```

### **Define Function**

```python
def income(wage: float, hours: int) -> float:
	return wage * hours + (hours > 40)*(hours-40)*.5*wage
```

### **Partially Apply & Map Function**

```python
hours = np.linspace(0, 80, 80)
f = partial(income, 20.0)
ys = list(map(f, hours))
```

### **Plot**

```python
plt.plot(hours, ys)
plt.title('Income', loc='left')
plt.xlabel("Hours Worked", size=14)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```
