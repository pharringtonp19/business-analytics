### **Import Packages**

```python
import jax
import jax.numpy as jnp
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import statsmodels.formula.api as smf
import seaborn as sb
```

### **Read In Data Set**

```python
df = pd.read_csv('/content/business-analytics/datasets/CASchools.csv')
```

```python
eqn = smf.ols('math ~ income', data = df)
results = eqn.fit()
xs = jnp.linspace(df['income'].min(), df['income'].max(), 100)
ys = results.params.values[0] + xs*results.params.values[1]
```

### **Plot**

```python
eqn = smf.ols('math ~ income', data = df)
results = eqn.fit()
xs = jnp.linspace(df['income'].min(), df['income'].max(), 100)
ys = results.params.values[0] + xs*results.params.values[1]

plt.scatter(df['income'], df['math'], color="#36454F")
plt.plot(xs, ys)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.xlabel('Size', size=14)
plt.title('Price', size=14, loc='left')
plt.show()
```

### **Seaborn**

```python
sb.regplot(x = 'income', y='math', data=df, color="#36454F", line_kws = {'color': 'blue'})
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.xlabel('Size', size=14)
plt.title('Price', size=14, loc='left')
plt.ylabel('')
plt.show()
```

```python
sb.regplot(x = 'income', y='math', data=df, color="#36454F", line_kws = {'color': 'blue'}, x_bins=range(0, 60, 5))
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.xlabel('Size', size=14)
plt.title('Price', size=14, loc='left')
plt.ylabel('')
plt.show()
```
