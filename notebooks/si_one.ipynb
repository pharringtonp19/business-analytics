### **Import Libraries**

```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

### **Linear Function**

```python
def f(x):
 return 0.01*x

xs = jnp.linspace(18, 80,100)
ys = jax.vmap(f)(xs)
plt.plot(xs, ys)
plt.xlabel('Age', size=14)
plt.title('Probability of Voting for a Specific Candidate', loc='left')
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```

### **Non-linear Function**

```python
def f(x):
  x = (x - 50)/10
  return jax.nn.sigmoid(x)

xs = jnp.linspace(18, 80,100)
ys = jax.vmap(f)(xs)
plt.plot(xs, ys)
plt.xlabel('Age', size=14)
plt.title('Probability of Voting for a Specific Candidate', loc='left')
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```
