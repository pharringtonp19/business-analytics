### **Import Libraries**

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
import math
```

### **Read In the Dataset**

```python
df = pd.read_csv('/content/business-analytics/datasets/sportsref_download.csv')
df.head()
```

### **Invariance**

```python
def abs_correlation(col1, col2, a, b, c, d):
  df['col1_temp'] = df[col1].map(lambda x: a*x + b)
  df['col2_temp'] = df[col2].map(lambda x: c*x + d)
  return abs(df['col1_temp'].corr(df['col2_temp']))

print(abs_correlation('PTS', 'AST', 1, 8, 3, 4))
print(abs_correlation('PTS', 'AST', -12, 2.5, 12, 0.75))
```

### **Create Numerical Encoding for Positions**

```python
my_dict = {v:k for k,v in enumerate(pd.unique(df['Pos']))}
df['Pos_num'] = df['Pos'].map(my_dict)
plt.scatter(df['Pos_num'], df['PTS'])
plt.title('Points', loc='left')
plt.xlabel("Position", size=14)
plt.show()
```

### **For Loop**

```python
results = []
for pos in pd.unique(df['Pos']):  #'pd.unique(df['Pos']) creates a list-like object of all the possible values in df['Pos']
	condition = df['Pos'] == pos
	new_df = df[condition]
	avg = new_df['PTS'].mean()
	results.append(avg)

plt.bar(pd.unique(df['Pos']), results)
plt.title('Average Points', loc='left')
plt.show()
```
